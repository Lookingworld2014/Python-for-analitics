- Вывести на экран первые 5 строк
- Создать новый признак Cpu_Company, который будет содержать только название фирмы, которая произвела CPU (Intel, AMD и т.д)
- Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя (128, 1024, 256 и т.д.)
- Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD, SDD и т.д.)
- Удалите признаки Memory и ScreenResolution

```python
import pandas as pd
```

```python
df = pd.read_csv('laptop_price.csv',encoding='cp1251')
```

```python
df.head()
```

```python
def fun(strcpu):
    return strcpu.split()[0]
```

```python
df['Cpu_Company'] = df['Cpu'].apply(fun)
```

```python
df[['Cpu','Cpu_Company']] #внешние 
#внутренние аргументы(объект нужного типа)
```

```python
df['Cpu_Company'].value_counts()
```

```python
df['Memory_Amount'] = 0
```

```python
df['Memory'].value_counts()
```

```python
def tbg(string):
    return int(float((string.split()[0][:-2])))
```

```python
tbg('1TB HDD ')
```

```python
tbg('256GB SSD +  1.0TB Hybrid  ')
```

```python
tbg('1.0TB HDD')
```

```python
df['Memory_Amount'] = df['Memory'].apply(tbg)
```

```python
def type1(stroka11):
    return stroka11.split()[1]
```

```python
df['Memory_Type'] = df['Memory'].apply(type1)
```

```python
type1('256GB SSD +  1.0TB Hybrid  ')
```

```python
df.head()
```

```python
df['Memory_Type'].value_counts()
```

```python
#Удалите признаки Memory и ScreenResolution
```

```python
df.drop(columns=['Memory', 'ScreenResolution', 'Norm'])
```

```python
df.drop(columns=['Memory', 'ScreenResolution', 'Norm'], inplace=True)
```
