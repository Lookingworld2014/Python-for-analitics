4.1 Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

    - Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
df = df.drop(columns=['sqft_basement','sqft_above',	'zipcode','lat','long','sqft_living15','sqft_lot15'])
```

```python
pricehous = df.corr(numeric_only=True)
pricehous = np.round(pricehous,2)
pricehous = np.abs(pricehous)
pricehous
```

```python
plt.figure()
sns.heatmap(pricehous, annot = True, linewidths=0.5 , cmap='coolwarm')
```

Благодаря матрице кореляций - мы видим, что больше всего приближены к "price" - колонки: "sqft_living", "grade" и "bathrooms"
